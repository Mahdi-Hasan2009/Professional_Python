```python
import matplotlib.pyplot as plt

students_name = ["Rike", "Farhan", "Nasit", "Arafat", "Himel", "Yamin", "Zarif", "Mahdi", "Spondon", "Adian"]
student_marks = [35, 30, 35, 33, 32, 31, 36, 37, 38, 39]
marks_perc = []
for x in student_marks:
  res = (x/50)*100
  marks_perc.append(res)
print(marks_perc)

# Line Chart
def marks_line_chart():
  plt.plot(students_name, student_marks)
  plt.title("Students Marks Graph")
  plt.xlabel("Students Name")
  plt.ylabel("Students Marks")
  plt.show()

marks_line_chart()

# Bar Chart
def percentage_bar_chart():
  plt.bar(students_name, marks_perc)
  plt.title("Students' Parcentage Graph")
  plt.xlabel("Student Names")
  plt.ylabel("Student Parcentage")
  plt.show()

percentage_bar_chart()
```

```python
import matplotlib.pyplot as plt

month = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
profit = [1210, 1200, 2500, 3000, 2750, 2598, 1234, 4321, 1245, 5897, 5412, 5123]
potato = [30, 40, 50, 67, 76, 52, 51, 56, 35, 34, 25, 85]
onion = [20, 10, 30, 40, 50, 60, 70, 80, 90, 100, 110, 120]

def line_chart_of_month_wise_profit():
  plt.plot(month, profit, marker = "o", color="r", linestyle="dashdot", linewidth = 3, mec = "b")
  plt.title("Profit Month Wise")
  plt.xlabel("Month")
  plt.ylabel("Profit")
  plt.show()

line_chart_of_month_wise_profit()

def bar_chart_of_months_against_potato():
  plt.bar(month, potato)
  plt.title("Month vs. Potato")
  plt.xlabel("Months")
  plt.ylabel("Potato")
  plt.show()

bar_chart_of_months_against_potato()

def bar_chart_of_months_against_onion():
  plt.bar(month, onion)
  plt.title("Month vs. Onion")
  plt.xlabel("Months")
  plt.ylabel("Onion")
  plt.show()

bar_chart_of_months_against_onion()
```
