1. Connecting with SQL

```python
import sqlite3 as Mahdi
conn = Mahdi.connect("basketball.sqlite")
print("Database connected Successfully")
```

2. All the Tables

```python
import pandas as pd
tables = pd.read_sql("""SELECT  * FROM sqlite_master WHERE type = 'table';""", conn)
tables
```

3. THE TABLES

```python
Player = pd.read_sql("""SELECT * FROM Player;""",  conn)
```

```python
Team = pd.read_sql("""SELECT * FROM Team""", conn)
```

6. OPERATIONS

```python
# AVERAGE
AVG = pd.read_sql("""SELECT AVG(year_founded) AS 'AVG OF YEAR' FROM Team;""", conn)
AVG
```

```python
# SUM
SUM = pd.read_sql("""SELECT SUM(is_active) AS 'SUM' FROM Player;""", conn)
SUM
```

```python
# ACTIVE PLAYERS
ACTIVE_PERSONALS = pd.read_sql("""SELECT * FROM Player WHERE is_active = '1';""", conn)
ACTIVE_PERSONALS
```

```python
# UNIQ STATES OF TEAM
STATE = pd.read_sql("""SELECT DISTINCT state FROM Team;""", conn)
STATE
```
